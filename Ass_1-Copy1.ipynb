```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# data upload
df = pd.read_csv(r'Assignment2-data.csv',header=0,index_col=0)
```

```python
# data and data features
print(df)
print(type(df))
print(df.shape)
```

```python
# choose color palette
colormap = sns.color_palette("Reds")

# set figure properties
fig, ax = plt.subplots(figsize=(10,10))
sns.set(font_scale=1.5)

# create heatmap
ax = sns.heatmap(df,
                 cmap=colormap,
                 annot=True,
                 fmt = ".4g",      
                 vmin=0,
                 vmax=100,
                 linewidths=0.2,
                 square=True,
                 ax = ax,
                 cbar_kws={"shrink":.8,"label": "% Obesity"})
ax.set_title('Obesity over the life course across age groups', 
             fontsize=22, 
             fontdict={},
             pad = 20)

#configure the x and y ticks
plt.xticks(rotation=0, fontsize="9", va="center")
plt.yticks(rotation=0, fontsize="11", va="center")

# set axes labels
ax.set_xlabel('Birth cohort', fontsize=14, labelpad = 6)
ax.set_ylabel('Age group', fontsize=14, labelpad = 6)

```

```python
# invert rows and cols
inv_df = df.iloc[::-1]
inv_df = inv_df[inv_df.columns[::-1]]
```

```python
# print inv_df
print(inv_df)
```
