```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("final.csv")
```

```python
# per rispondere alle mie domande mi basta avere gli elements, quindi elimino le rige relative ai components
print("initial df.shape:",df.shape)
df.drop(df[df.section == "components"].index, inplace=True)
print("post cut df.shape:",df.shape)
```

```python
# elimino le colonne che non mi servono più
df.drop('total_segment_score', axis=1, inplace=True)
df.drop('total_component_score', axis=1, inplace=True)
df.drop('total_element_score', axis=1, inplace=True)
df.drop('section', axis=1, inplace=True)
```

```python
# quali sono le categorie di program?
df['program'].value_counts()
```

```python
# creo una colonna caregory in cui sono presenti le categorie: ladies, men, pairs, ice_dance
# create a list of our conditions
conditions = [
    (df['program'] == "LADIES FREE SKATING") |
    (df['program'] == "LADIES SHORT PROGRAM") |
    (df['program'] == "JUNIOR LADIES FREE SKATING") |
    (df['program'] == "JUNIOR LADIES SHORT PROGRAM"),
    (df['program'] == "MEN FREE SKATING") |
    (df['program'] == "MEN SHORT PROGRAM") |
    (df['program'] == "JUNIOR MEN FREE SKATING") |
    (df['program'] == "JUNIOR MEN SHORT PROGRAM"),
    (df['program'] == "PAIRS FREE SKATING") |
    (df['program'] == "PAIRS SHORT PROGRAM") |
    (df['program'] == "JUNIOR PAIRS FREE SKATING") |
    (df['program'] == "JUNIOR PAIRS SHORT PROGRAM"),
    (df['program'] == "ICE DANCE FREE SKATING") |
    (df['program'] == "ICE DANCE SHORT PROGRAM") |
    (df['program'] == "JUNIOR ICE DANCE FREE SKATING") |
    (df['program'] == "JUNIOR ICE DANCE SHORT PROGRAM")
    ]

# create a list of the values we want to assign for each condition
values = ['LADIES', 'MEN', 'PAIRS', 'ICE_DANCE']

# create a new column and use np.select to assign values to it using our lists as arguments
df['category'] = np.select(conditions, values)

# display updated DataFrame
df.head()
```
